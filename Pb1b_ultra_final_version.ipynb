```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

#function defintion needed for integrator
def Hopf_model(Z,t): #two input arguments Z and t, Z=[x,y]
    
    #Given differential equations
    #a will be input later when using the integrator
    dxdt=a*Z[0]+Z[1]-Z[0]*(Z[0]**2+Z[1]**2)#Z[0]=x
    dydt=-Z[0]+a*Z[1]-Z[1]*(Z[0]**2+Z[1]**2)#Z[1]=y
    dZdt=[dxdt,dydt] #derivative of Z has two rows, derivative for x and derivative for y
    
    return dZdt #function returns the derivatives

a=10  #Choose value of a as positive
Z_initial1=[1,1] #initial x and y chosen
t=np.linspace(0,25,500) #time array
Z=odeint(Hopf_model,Z_initial1,t) #odeint integrator function used


x=np.array(Z[:,0]) #obtaining first column which is x from Z solved by odeint for different times using intial Z and the fucntion defined above
y=np.array(Z[:,1])#obtaining second column which is x from Z solved by odeint for different times using intial Z and the fucntion defined above

#theta=np.arctan(x/y)
#print(theta)


#Plotting 
fig,(a1)=plt.subplots(1)

a1.plot(t,Z[:,0],color='red',label='x')
a1.plot(t,Z[:,1],color='black',label='y')

a1.legend(loc='lower left')
a1.set_xlabel('t')
a1.set_ylabel('x or y')
a1.title.set_text('a=10, Initial: [1,1]')
plt.savefig('Pb1b_a=10_ultra_final_version.png')



#Again solving Z using odeint but this time as a<0
Z_initial2=[1,1]
a=-2
Z2=odeint(Hopf_model,Z_initial2,t)
fig,(a2)=plt.subplots(1)

a2.plot(t,Z2[:,0],'r-',color='red',label='x')
a2.plot(t,Z2[:,1],'b-',color='black',label='y')

a2.legend(loc='lower left')
a2.set_xlabel('t')
a2.set_ylabel('x or y')
a2.title.set_text('a=-2, Initial: [1,1]')
plt.savefig('Pb1b_a=-2_ultra_final_version.png')


#Again solving Z using odeint but this time for different intial conditions for a=-2 
Z_initial3=[4,3]
a=-2
Z3=odeint(Hopf_model,Z_initial3,t)

fig,(a3)=plt.subplots(1)
a3.plot(t,Z3[:,0],'r-',color='red',label='x')
a3.plot(t,Z3[:,1],'b-',color='black',label='y')
a3.legend(loc='lower left')
a3.set_xlabel('t')
a3.set_ylabel('x or y')
a3.title.set_text('a=-2, Initial: [4,3]')
plt.savefig('Pb1b_a=-2 AND Different intial conditions.png')



plt.show()
```
