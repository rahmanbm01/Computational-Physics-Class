```python
import matplotlib.pyplot as plt
#  FUNCTION DEFINITION
def logistic_calculate(r,x):
    #r and x are both passed by value, and cannot be updated within the function
    return r*x*(1-x)
    #this returns the new value of x.
# END OF FUNCTION DEFINITION


# PARAMETERS
max_iterations=51  # how many iterations to do
numprint=30	 #how many numbers to print to the screen
  # our choice for r.

# INITIALIZATION
x=0.01    # initial value for x


iter_list_r1=[0] #initializing list for number of successive iterations for r=2.00
iter_list_r2=[0] #initializing list for number of successive iterations for r=2.99
x_list_r1=[x] #initializing list for successive x values for r=2.00
x_list_r2=[x] #initializing list for for successive x values for r=2.99
r_list=[2.00,2.99] #list with r values


#iteratively update the value for x.
for r in r_list: #iterating over r values
    for iter in range(1, max_iterations):

        x=logistic_calculate(r,x);
        # x must be updated with the function by hand, since it was by value only.
        if r==r_list[0]: #if r=2.00, x_list_r1 appended
            x_list_r1.append(x)
            iter_list_r1.append(iter)
        if r==r_list[1]: #if r=2.99, x_list_r2 appended
            x_list_r2.append(x)
            iter_list_r2.append(iter)

    x=0.01 #using initial x given for r=2.99
    
#plotting commands
plt.plot(iter_list_r1,x_list_r1, marker=".", color='r')
    
plt.xlabel('iterations')
plt.ylabel('x')

plt.plot(iter_list_r2,x_list_r2, marker="*", color='b')
plt.show()

plt.savefig('Pb5_py.png')
```
